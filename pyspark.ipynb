```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
spark.conf.set("spark.sql.repl.eagerEval.enabled", True) # Property used to format output tables better
spark
```

```python
df=spark.read.csv('/content/mobile prediction.csv',header=True,inferSchema=True)
df
```

```python
train,test=df.randomSplit([.8,.2],seed=42)
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
vecassembler=VectorAssembler(inputCols=['ram'],outputCol='feature')
```

```python
trainvec=vecassembler.transform(train)
```

```python
trainvec.printSchema()
```

```python
trainvec.select('*').show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
le=LinearRegression(featuresCol='feature',labelCol='ram')
```

```python
model=le.fit(trainvec)
```
